# BASICS OF RECURRENT NEURAL NETWORKS
A recurrent neural network (RNN) is a type of artificial neural network which uses sequential data or time series data. In recurrent neural networks, the output of certain nodes affect the subsequent input to the same nodes. RNNs make use of internal states to process variable length sequences of inputs. RNNs are typically used for ordinal or temporal problems.

RNNs utilize a 'hidden state' that allows them to retain information about previous inputs, which are then used in the current inputs. This results in a temporal dependency in the model.

Each neuron in a RNN combines the input at each time step with the previous hidden states to produce an output and update the hidden state. Various gating mechanisms are used to decide what inputs go into a neuron and which ones do not.

RNNs share parameters across each layer of the network. While feedforward networks have different weights across each node, recurrent neural networks share the same weight parameter within each layer of the network. These weights are then adjusted through the process of backpropagation and gradient descent.

RNNs utilize an algorithm called Backpropagation Through Time (BPTT) to determine the gradients. BPTT is an update on the traditional backpropagation that's adjusted for sequential data. In BPTT, the errors are summed at each time step; while in traditional backpropagation, errors are not summed as parameters are not shared across each layer.

RNNs can handle variable-length sequences as inputs or produce variable-length sequences as outputs. They are useful for processing varying amounts of input data, i.e. you can have 1 input value or 5, it doesn't matter.

## Problem with traditional Recurrent Neural Networks
One of the main
shortcomings of vanilla RNN is that they fail to consider long-term dependencies that may be present in a sequence. A traditional RNN only processes one, or at most a very few inputs at a time.
<br>
e.g. Hi, how are you? I am ____

If we use a traditional RNN to process this sequence and autocomplete the sentence, it may fail to consider the words 'cat' or 'sitting'. So, the suggestion may be a word like 'John', which is accurate if we only consider the second sentence, but isn't appropriate when we consider the entire text.

On the other hand, using LSTM or GRU ensures that the context of the previous sentence isn't lost, and the resulting suggestion would be a word like 'good' or 'fine', which is appropriate given the context.


Another shortcoming of traditional RNNs is what is known as the "vanishing and exploding gradients" problem.

When training a recurrent neural network, sometimes the gradients that are used to update the neural networks can become very small or very large in magnitude. These problems are known as the 'vanishing' and 'exploding' gradients problem, respectively.

Vanishing gradients problem occurs when the gradients are very small. This causes the magnitude of the gradient to decrease exponentially, which can cause extremely small updates to the network parameters, causing slow learning.

Conversely, exxploding gradients problem occurs when the gradients are very large. This causes the magnitude of the gradient to increase exponentially during backpropagation, which can cause extremely large updates to the network parameters, which prevents convergence of the loss to a minima.

# GRAPH DATA WITH SEQUENCES - INTRO TO RGNNS

Graph Neural Networks are a subset of Neural Networks that are used to process data present in the form of graphs. They are similar to traditional neural networks, but the type of data they process is vastly different; traditional neural networks take in structured data as input, while GNNs can take graphs, which is a type of unstructured data, as input.

Traditional GNNs are useful to capture relationships among entities (like nodes and edges) in a graph by using message passing and aggregation mechanisms. However, they fail to take into account the sequential dependencies that may be present in a graph. This shortcoming is addressed by Recurrent Graph Neural Networks.

on many end-to-end learning tasks. However, they lack a high-level and intuitive spatio-temporal
structure. This spatio-temporal relationship can be modeled by a spatio-temporal graph, which can then be processed by neural networks.

Recurrent Graph Neural Networks (RGNNs) are neural network models that utilize the functionality of traditional Recurrent Neural Networks to process and analyze graph-structured data that possess sequential dependencies. It is an extension of Graph Neural Networks that facilitates the processing of sequential graph data.
Being a combination of two very powerful domains in Machine learning, this neural network model is very impactful for graph-structured datasets that exhibit sequential dependencies.
<br>
Applications of RGNNs can be found in a multitude of domains that require processing of data containing interconnected vertices; one very popular example being traffic prediction. Recurring incidents
such as rush hours or accidents can cause nonstationarity, making it difficult to forecast longterm. Road network is structured similar to a graph, because road networks are interconnected and may have weights corresponding to the traffic load, etc. When analyzing traffic data, an added dimensionality of  temporal dependency is added to the graph, resulting in a graph with temporal dependencies within the structure.



While traditional RNNs are also useful in analysis of such data, they only capture the temporal dependencies in the data, failing to take into account spatial dependencies that may exist (traffic flow in one segment of the road may be affected due to traffic flow in other roads). Conversely, GNNs factor in spatial dependencies but fail to consider temporal dependencies. RGNNs can capture both spatial and temporal characteristics of the data, making it much more useful than its elements.

# Graph LSTM and Graph GRU

Graph LSTM is a neural network architecture that extends the Long Short Term Memory (LSTM) model to handle structured data represented as graphs that possess sequential dependencies.

Much like traditional LSTM models, Graph LSTMs are capable of handling long-term dependencies within a sequence, and mitigate the vanishing & exploding gradients problem that may be prevalent in Recurrent GNNs.

The graph is first embedded into a vector, typically of a lower dimension. Then, the nodes in that graph is mapped into a sequence.

In Graph LSTM, the LSTM model is modified to accommodate both node attributes and the neighborhood information with the sequential data, which in the case of RGNNs is a sequence of nodes.


Graph LSTM implementation, one of many methods:
embed nodes into a low dimensional vector space, and use random walk based algorithm to map the nodes into a sequence. The transformation of nodes to a sequence can be approximated by any distribution, like the normal distribution.

The RNN architecture takes both node embeddings and node sequences as input, i.e. it takes in both the node representation, and the neighborhood information as input.

Then, the predictive function is given by:

$$Y_i = f\left(\sum_{s=1}^{k} w_s \sum_{j \in N_s(i)} H_j \cdot E\right)$$


where $$ \left(w_s \sum_{j \in N_s(i)} H_j \cdot E \right)$$
<br>
is the sum of node embeddings from the set of neighbors that are at a distance *S* to the center node *i*.


<br>

The loss function can be either Negative Log Loss or sum of cross-entropy over all nodes.

The state update equation is
$$ g'(h_t, x_t, N_{bt}) $$


<br>

Recall that the update equations for a traditional LSTM model are:

input gate *i* : $$
i_t = \sigma(W_{ii}e_{v\pi(t)} + W_{hi}h_{t-1} + b_i)
$$

Forget gate *f* :
$$
f_t = \sigma(W_{if}e_{v\pi(t)} + W_{hi}h_{t-1} + b_f)
$$

Output gate *o* :
$$
o_t = \sigma(W_{io}e_{v\pi(t)} + W_{ho}h_{t-1} + b_o)
$$



Cell state *g*, candidate cell state *c* and final output *h*:
$$
g_t = \tanh(W_{ig}e_{v\pi(t)} + W_{hg}h_{t-1} + b_g)
$$

$$
c_t = f_t \cdot c_{t-1} + i_t \cdot g_t
$$

$$
h_t = o_t \cdot \tanh(c_t)
$$


The same equation is modified in case of a Graph LSTM, wherein a term is added within the sigmoid function. This term is a representation of the neighborhood informtaion, and is of the form <br>
$$W_{nx}Nb_{v\pi(t)}\ $$

where $$W_{nx} $$ is the weight matrix associated with the gate, and $$Nb_{v\pi(t)}$$ is the neighborhood information concerning the central node.

<br>

Thus, the updated equations for Graph LSTM are:

$$
i_t = \sigma(W_{ii}e_{v\pi(t)} + W_{ni}Nb_{v\pi(t)} + W_{hi}h_{t-1} + b_i)
$$

$$
f_t = \sigma(W_{if}e_{v\pi(t)} + W_{nf}Nb_{v\pi(t)} + W_{hi}h_{t-1} + b_f)
$$

$$
o_t = \sigma(W_{io}e_{v\pi(t)} + W_{no}Nb_{v\pi(t)} + W_{ho}h_{t-1} + b_o)
$$




with the cell state update equations as follows:

$$
g_t = \tanh(W_{ig}e_{v\pi(t)} + W_{ng}Nb_{v\pi(t)} + W_{hg}h_{t-1} + b_g)
$$

$$
c_t = f_t \cdot c_{t-1} + i_t \cdot g_t
$$

$$
h_t = o_t \cdot \tanh(c_t)
$$


<br>
<br>
<br>

Here, $$W_{ix}$$ represents the corresponding weight matrices for the various gates.


## GRAPH GRU

Graph GRUs implement time gating mechanisms to address problems associated with long term dependencies.

Gated Graph Neural Networks use GRUs to reduce recurrence to a fixed number of steps.

$$
h_v^{(t)} = \text{GRU}\left(h_v^{(t-1)}, \sum_{u \in N(v)} W \cdot h_u^{(t-1)}\right)
$$


The graph weights *W* can be initialized to a matrix of shape *n* * *n*, where *n* is the number of graph embeddings.

$$
h_i^{(0)} = x_i \, || \, 0
$$


$$
m_i^{(l+1)} = \sum_{j \in N(i)} e_{j,i} \cdot \theta \cdot h_j^{(l)}
$$


$$
h_i^{(l+1)} = \text{GRU}(M_i^{(l+1)}, h_i^{(l)})
$$

# Gating Mechanisms for Gated Graph Neural Networks

Graph GRU - node, edge and time gating

The simplest gating mechanism is time gating. Time gating is an extension of input and forget gates of LSTM.
Time gating multiplies the input and the state by scalar gates $$\hat{q} \in [0, 1] $$ and $$\check{q} \in [0, 1] $$



The node and edge structure of the graph allows for other forms of gating, generally referred to as 'spatial gating'.
Spatial gating encompasses two forms of gating: Node gating and Edge gating.
Node gating - one input and one forget gate for each node of the graph.
Spatial gating strategies help encode long range spatial dependencies in graph processes.

Multiply input and state by diagonal matrices; so each nodal component of the signal is multiplied by a different value.

Edge gated GRNN - elementwise multiplication of the shift operator by gate matrices; $$\hat{q} \in [0, 1]^{N*N} $$ and $$\check{q} \in [0, 1] ^ {N*N} $$

separate gate for each edge, controls the amount of information transmitted across edges.

Parameters of input and forget gate operators are the outputs of GRNNs themselves.

# GCNs AND RGNNs

The architecture of an RGNN typically consists of two main components: a graph convolutional layer and a recurrent layer. The graph convolutional layer performs message passing and aggregation operations to gather information from neighboring nodes, capturing the local structural dependencies. The recurrent layer then updates the node representations based on the current state, the aggregated information, and the previous hidden state.
