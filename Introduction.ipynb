# **Introduction to Graphs**

A graph, in the most basic sense, is defined as a set of objects that are related to each other in some manner. It is a non-linear data structure that consists of ‘vertices’, also called ‘nodes’; and ‘edges’, also called ‘links’. Graphs are used to represent and analyze relationships between entities.

Graphs can be broadly classified into two types: Directed graphs and Undirected graphs.


* Directed graphs are graphs that have directed edges between nodes. They show an asymmetrical relationship between nodes.
*  Undirected graphs are graphs that contain undirected edges between nodes and show a symmetric relationship between nodes.
The edges of an undirected graph define a symmetric relation on the vertices, called the adjacency relation.
Symmetric relation - a type of binary relation that applies symmetrically (to both vertices in case of graphs)




## Formal definition of graphs
- An undirected graph is a pair G = (V, E) where V is a set of elements called vertices, and E is a set of unordered pairs of vertices, whose elements are called edges.
- A directed graph is a pair G = (V, E) where V is a set of elements called vertices, and E is a set of directed edges, which are ordered pairs of distinct vertices
- In case of multiple edges, a directed graph is an ordered triple G = (V, E, ϕ) comprising: <br>
  * V - a set of vertices <br>
  * E - a set of edges <br>
  * Φ - an incidence function mapping every edge to an ordered pair of vertices <br>

## Key Terms



1.  Vertices: Vertices are the fundamental units of the graph. Sometimes, vertices
are also known as vertex or nodes. Every node/vertex can be labeled or unlabelled.

2. Edges: Every connecting line between two vertices are called edges. It can be an ordered pair of nodes in a directed graph. Edges can connect any two nodes in any possible way. There are no rules. Sometimes, edges are also known as arcs. Edges may be labeled or unlabelled.
  * Undirected Edge: It is a bi-directional edge.
  * Directed Edge: It is a unidirectional edge.
  * Weighted Edge: An edge associated with value in it.
  * Loops - edges that connect a vertex to itself

3.   Order of a graph - the number of vertices present in the graph.

4.    Size of a graph - the number of edges in the graph. It may also be the total number of vertices and edges in the graph, in some cases

5.   Degree of valency: Degree of valency of a node in a graph is the number of edges incident on it.
  * Indegree: Total number of incoming edges connected to the vertex.
  * Outdegree: Total number of outgoing edges connected to a vertex.

6.  Neighbors - Two nodes that are connected by an edge are called neighbors, and the nodes are said to be adjacent.

7.  Self-loop: If two endpoints coincide with each other, called self-loop.

# Representing data in graphs

## Edge lists
One of the ways of representing graphs is by using lists, or an array of $|E|$ edges. We call this list an edge list. In this implementation, the underlying data structure for keeping track of all the nodes and edges is a single list of pairs

To represent an edge, we have an array of two vertex numbers, or an array of objects containing the vertex numbers of the vertices that the edges are incident on. If edges have weights, we can add either a third element to the array or more information to the object, giving the edge's weight. Since each edge contains just two or three numbers, the total space complexity for an edge list is $Θ(E)$.

## Adjacency Matrix

A convenient way to represent graphs is through an adjacency matrix $A ∈
R (|V|×|V|)$. To represent a graph with an adjacency matrix, we order the nodes
in the graph so that every node indexes a particular row and column in the
adjacency matrix. We can then represent the presence of edges as entries in this matrix: $A[u, v] = 1$ if $(u, v) ∈ E$ and $A[u, v] = 0$ otherwise.
This means that if edges $u$ & $v$ have an edge between them, then that corresponding entry in the matrix is 1, otherwise it is 0.


* The rows and columns of the matrix represent the nodes in the graph, and each entry in the matrix indicates whether there is an edge between the two nodes. Typically, a value of 1 indicates the presence of an edge, while a value of 0 indicates the absence of an edge. This type of graph is useful for representing graphs with a small number of nodes and a dense set of edges. *$A[u][v]$* is **$1$** if there is an edge from $u$ to $v$, otherwise it is **$0$**. (In weighted graphs, if A is numeric valued, then *$A[u][v]$* would be the weight of the edge $(u,v)$.)

- For undirected graphs, the adjacency matrix is always symmetric, meaning *$A[u, v] = A[v, u]$* for all vertices u and v, because uv and vu are just different names for the same edge, and the diagonal entries A[u, u] are all zeros. For directed graphs, the adjacency matrix may or may not be symmetric, and the diagonal entries may or may not be zero. # Machine Learning and Graph Neural Networks


Graph Neural Networks (GNN) are a class of artificial neural networks that process data that can be represented as graphs.
Graphs can be used to represent data in a way that preserves the structure and relationships between different data points, which can be useful for tasks like classification, clustering, and anomaly detection.

Existing Neural Networks are more successful for structured data types. However, graphs are inherently unstructured, and conventional Neural Networks don't work in them. They don’t have a fixed structure or ordering. The existing Neural Networks can be interpreted as Graph Neural Networks that operate on suitably defined graphs. Convolutional Neural Networks operate on multi-dimensional graphs with a fixed shape. They can be called a ‘grid graph’. On the other hand, Recurrent Neural Networks operate on complete graphs where each node represents words in a sentence.


Graph Neural Networks are a subclass of Neural Networks that operate directly on graph-structured data. They can learn from and reason about complex graph structures, and they work based on the concept of ‘message passing’. At each layer of the network, each node aggregates information from its neighboring nodes and updates its own representation based on this information. This process is repeated for multiple layers, allowing the network to capture increasingly complex patterns and relationships within the graph.
Due to this, Graph Neural Networks are also known as Message Passing Neural Networks.


GNNs are used to solve problems in situations that can be represented as graphs, like social media, transportation, molecular biology, etc.


There are many different types of machine learning algorithms that work with graphs and networks, including graph neural networks, random walks, community detection algorithms, and more. These algorithms often involve manipulating or analyzing the structure of the graph in order to extract useful information or patterns from the data.


There are many different types of GNNs, each with its own unique architecture and method of message passing. Some of the most popular types of GNNs include graph convolutional networks (GCNs), graph attention networks (GATs), and message passing neural networks (MPNNs). These networks have been applied to a wide range of tasks, from predicting protein-protein interactions to identifying clusters of similar individuals in social networks.

# Types of Graph Neural Network tasks


## Node Classification
- Suppose we have a large social network dataset with millions of users, and we suspect that a significant number of these users are actually bots. Identifying these bots is important for various reasons, such as preventing them from violating the social network's terms of service or avoiding advertising to them. Manually examining each user to determine if they are a bot would be impractical, so we need a model that can classify users as a bot or not given only a small number of manually labeled examples.

- This is where node classification comes in. The goal of node classification is to predict the label associated with all nodes in a graph, given only the true labels of a small subset of nodes in a training set. In this scenario, the label of a node would be either "bot" or "not bot."
- Node classification is a popular machine learning task on graph data, especially in recent years. Besides social networks, other examples of node classification include classifying the function of proteins in the interactome and classifying the topic of documents based on hyperlink or citation graphs.
- However, node classification is different from standard supervised classification because the nodes in a graph are not independent and identically distributed (i.i.d.). In supervised machine learning models, we assume that each data point is statistically independent from all the other data points, but this is not the case with graphs because nodes are interconnected. Therefore, node classification models must explicitly leverage the connections between nodes in the graph.
- One way to do this is through the concept of homophily, which is the tendency for nodes to share attributes with their neighbors in the graph. For example, people tend to form friendships with others who share the same interests or demographics. Based on the notion of homophily, we can build machine learning models that try to assign similar labels to neighboring nodes in a graph.
- Node classification is often referred to as semi-supervised learning because when training the model, we usually have access to the full graph, including all the unlabeled (e.g., test) nodes. The only thing we are missing is the labels of test nodes. However, we can still use information about the test nodes (e.g., knowledge of their neighborhood in the graph) to improve our model during training.
<br>


## Relation Prediction
- As we have gone through node classification, it is useful for inferring information about a node based on its relationship with other nodes on the graph. So, what about the missing relationship information, we need to tackle those.
- This task is referred to as link prediction , graph completion, and relational inference based on a specific application domain. Relation prediction is one of the more popular machine learning tasks with graph data and has many real- world applications like recommending content to users in social platforms, predicting drug side-effects. These tasks can be viewed as special cases of relation prediction.
- The standard setup for relation prediction is that we are given a set of nodes and an incomplete set of edges between these nodes. Our goal is to use this partial information to infer the missing edges.
- The complexity of the task depends on the type of graph data being examined. Simple graphs, such as social networks, can be tackled with simple heuristics based on the number of neighbors two nodes share, while more complex multi-relational graphs, such as biomedical knowledge graphs, require complex reasoning and inference strategies. Relation prediction can be both supervised and unsupervised and requires inductive biases specific to the graph domain. There are different variants of relation prediction, including settings where predictions are made over a single, fixed graph and settings where relations must be predicted across multiple disjoint graphs.
<br>


## Clustering and community detection
- Above discussed node classification and relation prediction, these two tasks are the graph analogues of supervised learning. Community detection, on the other hand, is the graph analogue of unsupervised clustering.
- When we look at a collaboration graph of researchers based on co-authored papers, we expect to see clusters of nodes that are more likely to collaborate with each other due to factors such as research area, institution, or demographics. This is called community structure and is the focus of community detection. The goal is to identify these communities in the graph, even if they are not explicitly labeled, and this has applications in identifying functional modules in genetic interaction networks and detecting fraudulent groups in financial transaction networks.
<br>


## Graph classification, regression, and clustering
- Machine learning on graphs includes applications such as classification, regression, and clustering over entire graphs. For example, predicting toxicity of a molecule or detecting malicious computer programs using graph-based representations. In these graph classification or regression applications, we seek to learn over graph data, but instead of making predictions over the individual components of a single graph (i.e., the nodes or the edges), we are instead given a dataset of multiple different graphs and our goal is to make independent predictions specific to each graph.
- Graph regression and classification are the most straightforward supervised learning analogues, while graph clustering is an extension of unsupervised clustering. The challenge in these tasks is defining useful features that consider the relational structure within each graph.

```python
import networkx as nx
```

```python
H = nx.DiGraph()

#adding nodes
H.add_nodes_from([
    (0, {"color":"blue", "size" :250}),
    (1, {"color":"yellow", "size": 400}),
    (2, {"color":"orange", "size": 150}),
    (3, {"color":"red", "size":600})
])

#adding edges
H.add_edges_from([
    (0,1),
    (1,2),
    (1,0),
    (1,3),
    (2,3),
    (3,0)
])

node_colors = nx.get_node_attributes(H, "color").values()
colors = list(node_colors)
node_sizes = nx.get_node_attributes(H, "size").values()
sizes = list (node_sizes)

#Plotting Graph
nx.draw(H, with_labels=True, node_color=colors, node_size=sizes)
```

```python
#converting to undirected graph
G = H.to_undirected()
nx.draw(G, with_labels=True, node_color=colors, node_size=sizes)
```
